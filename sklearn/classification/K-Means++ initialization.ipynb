```python
from sklearn.cluster import kmeans_plusplus
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate sample data
n_samples = 4000
n_components = 4
```

```python
# 绘制斑点图
# X 为二维坐标的点 (4000, 2)
# y_true 为标签，四组分[0, 1, 2, 3]
X, y_true = make_blobs(
    n_samples=n_samples, centers=n_components, cluster_std=0.60, random_state=0
)
print(f'the shape of origin X is {np.shape(X)}')
# change order for every point
X = X[:, ::-1]
print(f'the shape of transformed X is {np.shape(X)}')
print(f'the first element of transformed X is {X[0]}')
print(f'the type of transformed X is {type(X)}')

# Calculate seeds from kmeans++
# centers_init: ndarray of shape (n_clusters, n_features)
# indices: ndarray of shape (n_clusters,)
centers_init, indices = kmeans_plusplus(X, n_clusters=4, random_state=0)
print(f'\nthe shape of centers_init is {np.shape(centers_init)}')
print(f'the shape of indices is {np.shape(indices)}')

# Plot init seeds along side sample data
plt.figure(1)

colors = ["#4EACC5", "#FF9C34", "#4E9A06", "m"]

for k, col in enumerate(colors):
    # extract the data with the same label and assign them a unnique color
    # s: the size of the dot
    cluster_data = y_true == k
    print(f'\nthis is the {k} round')
    print(f'the shape of cluster_data is {np.shape(cluster_data)}')
    print(f'the first element of cluster_data is {cluster_data[0]}')
    print(f'the shape of X[cluster_data, 0] is {np.shape(X[cluster_data, 0])}')
    print(f'the first element of X[cluster_data, 0] is {X[cluster_data, 0][0]}')
    plt.scatter(X[cluster_data, 0], X[cluster_data, 1], c=col, marker=".", s=10)

# draw 4 all centers
plt.scatter(centers_init[..., 0], centers_init[..., 1], c="r", s=60)
plt.title("K-Means++ Initialization")
plt.xticks([])
plt.yticks([])
plt.show()
```
